```python
def sublists(n):
    numset = []
    for i in range(13,len(n)):
        subset = []
        slicedval = n[i-13:i]
        subset.append(slicedval)
        numset += subset
    return numset
```

```python
def multlist(n):
    totallist = sublists(n)
    product = []
    for sub in totallist:
        val = 1
        for num in sub:
            val *= int(num)
        product.append(val)
    return max(product)
```
